```python
from diffusers import StableDiffusionPipeline
import torch

#load the stable diffusion model
model_id ='CompVis/stable-diffusion-v1-4'
pipe = StableDiffusionPipeline.from_pretrained(model_id,
                                               torch_dttype=torch.float16)

#move the model to cpu if available
device = 'cuda' if torch.cuda.is_available() else 'cpu'
pipe = pipe.to(device)
```

```python
#move the mmodel device if availlable
device = 'cuda' if torch.cuda.is_available() else 'cpu'
pipe =pipe.to(device)
```

```python
#Define your text prompt
prompt = 'Flying cars soar over a futuristic cityscape at sunset '
#Generaate the image
with torch.autocast('cuda'): #use autocaast for mixed precision (faster on GPU)
  image = pipe(prompt).images[0]

  #save the image

image.save('generated_image.png')
print('image saved as generated_image.png')
```
