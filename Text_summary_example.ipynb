```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
```

```python
#load a pretraind model and tokenizer
model_name = "facebook/bart-large-cnn"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
#Input text to summarize

text = 'Virat kohli is in india international cricketer known for his exceptional batting skills and he has palyed for the india team in all format and is a former captain kohli is of referred to as king kohli and chase master due to bais remarkable ability to chase down'
```

```python
#input text
tokenizer.encode('summarize', text, return_tensors='pt',max_length=512, truncation=True)
```

```python
#Generate summary
inputs = tokenizer.encode('summarize', text, return_tensors='pt',max_length=512, truncation=True)
summary_ids = model.generate(inputs, max_length=50,min_length=25,length_penalty=2.0,num_beams=4)
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
```

```python
print(summary)
```
