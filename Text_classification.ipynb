```python
from transformers import pipeline

```

```python
#load the spam detection pipeline
spam_classifier = pipeline('text-classification',model='distilbert-base-uncased-finetuned-sst-2-english')
```

```python
pipeline('text-classification', model='distilbert-base-uncased-finetuned-sst-2-english')
```

```python
texts= [
    'Congratulations you ve won 500 INR Amazon gift card. click here to claim now.'
    'Hi Auta lets have a meeting tommorrow at 12PM'
    'URGENTS: Your gmail account has ben compromised. click here to secure it.'
]
results = spam_classifier(texts)
label_mapping ={'negative ': 'Spam', 'neural': 'Not spam', 'positive': ' Not spam'}

for result in results:
    label = label_mapping[result['label']]
    label = result['label']
    score = result['score']
    print(f'Label:(label), confidance:(score:.4f)')
```
