```python
def add_number(x,y):
  return x+y
```

```python
add_number(50,30)
```

```python
interface = gr.Interface(fn=add_number,
                         inputs=[gr.Number(),gr.Number()],
                         outputs= [gr.Number()])
```

```python
interface.launch()
```

```python
def sub_number(x,y):
  return x-y
```

```python
interface =gr.Interface(fn=sub_number,
                        inputs=[gr.Number(), gr.Number()],
                        outputs=gr.Number())
```

```python
interface.launch()
```

```python
def reverse_text(input_text):
  return input_text[::-1]
```

```python
reverse_text ('i like python')
```

```python
interface = gr.Interface(fn=reverse_text,
                        inputs =gr.Text(),
                         outputs=gr.Text())
```

```python
interface.launch()
```

Gradio Sliders

```python
def Slider_example(value):
  return f'slider current value is: {value}'
```

```python
Slider_example(20)
```

```python
interface = gr.Interface(fn=Slider_example,
                         inputs=gr.Slider(minimum=0,maximum=100,step=1),
                          outputs=gr.Text())
interface.launch()
```

```python
#Gradio image

from PIL import Image

def convert(image_path):
  #load the image
  image = Image.open(image_path)
  #convert yythe image into graysclae format
  return image.convert('L')
```

```python
convert ('/content/outreachy.PNG')
```

```python
iface=gr.Interface(fn=convert,
                   inputs=gr.Image(type='filepath'),
                   outputs=gr.Image())

iface.launch()
```

```python

#Gradio JSON


def number_details(number):
  details={
      'original':number,
      'squared':number**2,
      'sqrt': number**0.5,
      'is_even': number%2 ==0
  }

  return details


number_details(20)
```

```python
iface = gr.Interface(fn=number_details,
                     inputs=gr.Number(),
                     outputs=gr.JSON())
iface.launch()
```

```python
#Gradio label
def classify_number(number):
  if number > 0:
    return 'positive'
  elif number < 0:
   return 'negative'
  else:
    return 'zero'


classify_number(-1)
```

```python

iface =gr.Interface(fn=classify_number,
                    inputs=gr.Number(),
                    outputs=gr.Label())
iface.launch()
```

```python
with gr.Blocks() as demo:
  with gr.Row():
    with gr.Column():
      text1 = gr.Text(value='Row 0 col 0 -comp1')
      text2 = gr.Text(value='Row 0 col 0 -comp2')

    with gr.Column():
      text3 = gr.Text(value='Row 0 Col 1')

    with gr.Row():
      text4 =gr.Text(value ='button row')

demo.launch()
```

```python
with gr.Blocks() as demo:
  with gr.Row():
    with gr.Column(scale=2):
      text1 = gr.Text(value='Row 0 Col 0 - comp1')
      text2 = gr.Text(value ='Row 0 Col 0 - comp2')
    with gr.Column(scale=1):
      text3 =gr.Text(value='Row 0 Col 1')
    with gr.Row():
      text4 =gr.Text(value='Botton Row')
demo.launch(share=True)
```

```python
with gr.Blocks() as demo:
  with gr.Tab('Tab one'):
    with gr.Row():
      with gr.Column():
        text1 =gr.Text(value ='Row 0 Col0 -comp1')
        text2 =gr.Text(value='Row0 Col0 -comp2')

        with gr.Column():
          text3 = gr.Text(value ='Row0 Col0 -comp1')

        with gr.Row():
          text4 = gr.Text(value='Botton row')

    with gr.Tab('Tab two'):
      with gr.Row():
        gr.Text('welcome to the new tab!')
demo.launch()
```

```python
#Gradio Accordion
with gr.Blocks() as demo:
  gr.Label('Label here')
with gr.Accordion('Accordion here', open=True):
  gr.Image()

demo.launch()

```

```python
##Gradio Buttons
with gr.Blocks() as demo:
  gr.Button('Button comp one')
  gr.Button('Button comp two')
  gr.Image()

demo.launch(share=True)
```

```python
def multiply(x,y):
  return x *y
```

```python
multiply(10,5)
```

```python
with gr.Blocks() as app:
  with gr.Row():
    x_slider =gr.Slider()
    y_slider =gr.Slider()

  with gr.Row():
    result = gr.Text()
    y_slider.change(fn=multiply,
                    inputs=[x_slider, y_slider],
                    outputs=result)

app.launch(share =True)
```

```python
with gr.Blocks() as app:
  with gr.Row():
    x_slider =gr.Slider(label='x')
    x_slider =gr.Slider(label='y')
  with gr.Row():
    result =gr.Text()

  x_slider.change(fn=multiply,
                  inputs=[x_slider,y_slider],
                  outputs=result)
  y_slider.change(fn=multiply,
                  inputs=[x_slider,y_slider],
                  outputs=result)
app.launch(share=True)
```
